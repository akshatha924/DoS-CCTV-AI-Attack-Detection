```python
# Step 8.2: Import Libraries for EDA

import pandas as pd          # For handling datasets
import matplotlib.pyplot as plt  # For plotting graphs
import seaborn as sns        # For advanced visualizations

# Optional: make plots look nicer
sns.set(style="whitegrid")

```

```python
# Step 8.3: Load the network traffic dataset
data = pd.read_csv("../dataset/network_traffic.csv")  # Path to your CSV

# Display first 5 rows
data.head()
```

```python
# Step 8.4: Visualize Packet Rate (Normal vs DoS)

plt.figure(figsize=(6,4))
sns.barplot(x="label", y="packet_rate", data=data)
plt.xticks([0,1], ['Normal', 'DoS Attack'])
plt.title("Packet Rate: Normal vs DoS Attack")
plt.ylabel("Packets per second")
plt.show()
```

```python
# Step 8.5: Visualize Byte Rate (Normal vs DoS)

plt.figure(figsize=(6,4))
sns.boxplot(x="label", y="byte_rate", data=data)
plt.xticks([0,1], ['Normal', 'DoS Attack'])
plt.title("Byte Rate Distribution")
plt.ylabel("Bytes per second")
plt.show()
```

```python
# Encode 'protocol' column to numeric
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data['protocol'] = encoder.fit_transform(data['protocol'])

# Check first 5 rows to confirm
data.head()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```
